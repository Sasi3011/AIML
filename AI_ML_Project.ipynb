```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_excel("assets/Smart_Fertilizer_Datasett.xlsx")

# Categorical & numeric columns
categorical_cols = ["Region", "Crop_Type", "Soil_Type", "Usage_Timing"]
numeric_cols = ["Soil_pH", "Soil_Nitrogen", "Soil_Phosphorus",
                "Soil_Potassium", "Organic_Carbon",
                "Plant_Age_Weeks", "Rainfall_mm", "Temperature_C"]

# Encode categorical variables
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# Scale numeric features
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Check missing values
print("Missing Values:", df.isnull().sum())
```
